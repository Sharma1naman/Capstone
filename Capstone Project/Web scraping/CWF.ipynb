# Stars

```python
director = soup.find_all("p", class_ = '')
director1 = []
stars=[]

finaldirector = []
row=[]

star1=[]
star2=[]
star3=[]
star4=[]
for i in director:
    row.clear()
    for j in i.find_all():
        
        if(j.name=="span"):
            c=1
            for element in j.next_siblings:
                if element.name=='a':
                    if c==1:
                        star1.append(element.text)
                    elif c==2:
                        star2.append(element.text)
                    elif c==3:
                        star3.append(element.text)
                    elif c==4:
                        star4.append(element.text)
                    
                    c=c+1
```

# Genre

```python
genre = soup.find_all("p", class_="text-muted")
genre11=[]
row=[]

genre1=[]
genre2=[]
genre3=[]

for i in genre:
    c=1
    row.clear()
    for j in i.find_all("span", class_="genre"):
        genre11.append(j.text)
        row.append(j.text)
    #elements=row.split(',')
    for x in row:
        #print(x)
        
    
        if len(x.split(','))==1:
            genre1.append(x.split(',')[0].split('\n')[1])
            genre2.append("NaN")
            genre3.append("NaN")
        elif len(x.split(','))==2:
            genre1.append(x.split(',')[0].split('\n')[1])
            genre2.append((x.split(',')[1]).strip())
            genre3.append("NaN")
        else:
            genre1.append(x.split(',')[0].split('\n')[1])
            genre2.append((x.split(',')[1]).strip())
            genre3.append(x.split(',')[2].strip())
        
genre1
```

# Vote

```python
gross = soup.find_all("p", class_="sort-num_votes-visible")

Gross = []
vote=[]

for i in gross:
    if(len(i.find_all("span", {'name':'nv'})) == 2):
        c=1
        for j in i.find_all("span", {'name':'nv'}):
            if(c==1):
                vote.append(j.text)
            else:
                Gross.append(j.text)
            c=c+1
            
    elif(len(i.find_all("span", {'name':'nv'})) == 1):
        
        for j in i.find_all("span", {'name':'nv'}):
        
                vote.append(j.text)
            
                Gross.append("NaN")
            
            
vote
```
