# create database

```python
import sqlite3
```

```python
file = "namanfinal.db"
  
conn = sqlite3.connect(file) 
print("Database created and connected.") 
```

# create table 1

```python
cursor_obj = conn.cursor()
 
# Drop the GEEK table if already exists.
cursor_obj.execute("DROP TABLE IF EXISTS Table1")
 
# Creating table
table = ''' CREATE TABLE Table1 (
               Sno INT PRIMARY KEY NOT NULL,
               MovieName VARCHAR(255),
               Director1 VARCHAR(255),
               Director2 VARCHAR(255),
               Director3 VARCHAR(255),
               Duration VARCHAR(255),
               YearNo VARCHAR(255),
               Rating VARCHAR(255),
               MetaScore VARCHAR(255)
            ); '''
 
cursor_obj.execute(table)
 
print("Table is Ready")
```

# create table 2

```python
cursor_obj.execute("DROP TABLE IF EXISTS Table2")
 
# Creating table
table = ''' CREATE TABLE Table2 (
               MovieName VARCHAR(255),
               Star1 VARCHAR(255),
               Star2 VARCHAR(255),
               Star3 VARCHAR(255),
               Star4 VARCHAR(255),
               Votes VARCHAR(255),
               Genre1 VARCHAR(255),
               Genre2 VARCHAR(255),
               Genre3 VARCHAR(255),
               Gross_Collection VARCHAR(255),
               Popularity VARCHAR(255),
               Certification VARCHAR(255)
            ); '''
 
cursor_obj.execute(table)
 
print("Table is Ready")
```

# Insert Records In Table

```python
import csv
# reading data from the CSV file
with open('first.csv') as f:
    reader = csv.reader(f)
    data = list(reader)

data.pop(0)     
for row in data:
   cursor_obj.execute("INSERT INTO Table1 (Sno,MovieName,Director1,Director2,Director3,Duration,YearNo,Rating,MetaScore) values (?, ?, ?, ?, ?, ?, ?, ?, ?)", row)

print("records saved successfuly")
```

# Join

# Query1

```python
import pandas as pd
            
statement = '''
    SELECT Table1.Sno,Table1.MovieName,Table1.Director1,Table1.Director2,Table1.Director3,Table1.Duration,Table1.YearNo,
    Table1.Rating,Table1.MetaScore,Table2.Star1,Table2.Star2,Table2.Star3,Table2.Star4,Table2.Votes,Table2.Genre1,Table2.Genre2,
    Table2.Genre3,Table2.Gross_Collection,Table2.Popularity,Table2.Certification
    
    FROM Table1 INNER JOIN Table2
    ON Table1.MovieName = Table2.MovieName;
    
    '''
  
#cursor_obj.execute(statement)
data = pd.read_sql(statement, conn)
data
```

# Query 2 

```python
import pandas as pd
            
statement = '''
    SELECT Table1.MovieName,Table1.Director1,Table1.Director2,Table1.Director3,
    Table1.Rating, Table2.Gross_Collection
    
    FROM Table1 INNER JOIN Table2
    ON Table1.MovieName = Table2.MovieName
    and 
    (Table2.Genre1='Action' or Table2.Genre2='Action' or Table2.Genre3='Action')
    ;
    '''
  
#cursor_obj.execute(statement)
data = pd.read_sql(statement, conn)
data
```

# Query 3

```python
import pandas as pd
            
statement = '''
    SELECT Table1.Sno,Table1.MovieName,Table1.Director1,Table1.Director2,Table1.Director3,Table1.Duration,Table1.YearNo,
    Table1.Rating,Table1.MetaScore,Table2.Star1,Table2.Star2,Table2.Star3,Table2.Star4,Table2.Votes,Table2.Genre1,Table2.Genre2,
    Table2.Genre3,Table2.Gross_Collection,Table2.Popularity,Table2.Certification
    
    FROM Table1 INNER JOIN Table2
    ON Table1.MovieName = Table2.MovieName
    and
    Table2.Gross_Collection NOT IN ('NaN')
    order by  cast(replace(replace(Table2.Gross_Collection,'M',''),'$','') as INTEGER) DESC LIMIT 1
  
    
    '''
  
#cursor_obj.execute(statement)
data = pd.read_sql(statement, conn)
data
```

# Query 4

```python
import pandas as pd
            
statement = '''
    SELECT Table1.Sno,Table1.MovieName,Table1.Director1,Table1.Director2,Table1.Director3,Table1.Duration,Table1.YearNo,
    Table1.Rating,Table1.MetaScore,Table2.Star1,Table2.Star2,Table2.Star3,Table2.Star4,Table2.Votes,Table2.Genre1,Table2.Genre2,
    Table2.Genre3,Table2.Gross_Collection,Table2.Popularity,Table2.Certification
    
    FROM Table1 INNER JOIN Table2
    ON Table1.MovieName = Table2.MovieName
    where cast(Table1.Rating as DECIMAL(10,2)) = (select max(cast(Table1.Rating as DECIMAL(10,2))) from Table1);
      '''
  
#cursor_obj.execute(statement)
data = pd.read_sql(statement, conn)
data
```

# Query 5

```python
import pandas as pd
            
statement = '''
    SELECT Table1.Sno,Table1.MovieName,Table1.Director1,Table1.Director2,Table1.Director3,Table1.Duration,Table1.YearNo,
    Table1.Rating,Table1.MetaScore,Table2.Star1,Table2.Star2,Table2.Star3,Table2.Star4,Table2.Votes,Table2.Genre1,Table2.Genre2,
    Table2.Genre3,Table2.Gross_Collection,Table2.Popularity,Table2.Certification
     
    
    FROM Table1 INNER JOIN Table2
    ON Table1.MovieName = Table2.MovieName
    and
    cast(Table1.Rating as DECIMAL(10,2)) = (select min(cast(Table1.Rating as DECIMAL(10,2))) from Table1)
    and
    cast(replace(replace(Table2.Gross_Collection,'M',''),'$','') as DECIMAL(10,2)) >0
    order by
    cast(replace(replace(Table2.Gross_Collection,'M',''),'$','') as DECIMAL(10,2)) limit 1
    ;
    
    '''
    
#cursor_obj.execute(statement)
data = pd.read_sql(statement, conn)
data
```
